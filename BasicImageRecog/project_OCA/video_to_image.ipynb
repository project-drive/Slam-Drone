```python
import cv2 
import os 
  
# Read the video from specified path 
cam = cv2.VideoCapture("fan_3ns.mp4") 
  
try: 
    if not os.path.exists('fan_3'): 
        os.makedirs('fan_3') 
  
except OSError: 
    print ('Error: Creating directory of data') 

# frame 
currentframe = 317
while(True): 
      
    # reading from frame 
    ret,frame = cam.read() 
  
    if ret:  
        name = './fan_3/fan_' + str(currentframe) + '.jpg'
        print ('Creating...' + name) 
  
        # writing the extracted images 
        cv2.imwrite(name, frame) 
        currentframe += 1
    else: 
        break

cam.release() 
cv2.destroyAllWindows()
```
